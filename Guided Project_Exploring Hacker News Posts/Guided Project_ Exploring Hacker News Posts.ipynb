# Analyzing the submissions of posts to conclude which post comes out to be most popular on Hacker News, a extremely popular site in technology and startup circles.

> You can find the data set [here](<https://www.kaggle.com/hacker-news/hacker-news-posts>)
> Below we have provided an overview of our data set structure.

|**Header**|**Description**|
|:----|:----|
|id|The unique identifier from Hacker News for the post|
|title|The title of the post|
|url|The URL that the posts links to, if it the post has a URL|
|num_points|The number of points the post acquired, calculated as the total number of upvotes minus the total number of downvotes|
|num_comments|The number of comments that were made on the post|
|author|The username of the person who submitted the post|
|created_at|The date and time at which the post was submitted|

> In our this project we are specifically interested in only two 
> types of post `Ask HN` and `Show HN`. So we will slice our 
> data set into this two specific category. Then code to find out 
> which category receive more comments on average. And why? And 
> then code to analyse whether the post created in the category 
> with more comments on average, gets popular because it is being 
> post on a certain time period of the day.


```python
from csv import reader
```

```python
# open and read our data set
openfile = open('hacker_news.csv')
readfile = reader(openfile)
hn = list(readfile)
print(hn[:5])
```

```python
#separate header
header = hn[0]
print(header)
```

```python
# delete header from our data set
del hn[0]
```

```python
print(hn[:5])
```

```python
# separating posts beginning with Ask HN and Show HN
ask_posts = []
show_posts = [] 
other_posts = []

for lst in hn:
    title = lst[1]
    if title.lower().startswith('ask hn'):
        ask_posts.append(lst)
    elif title.lower().startswith('show hn'):
        show_posts.append(lst)
    else :
        other_posts.append(lst)

print(len(ask_posts))
print(len(show_posts))
print(len(other_posts))
print(ask_posts[:5])
print(show_posts[:5])
print(other_posts[:5])
```

```python
# code to find whether ask posts or show posts received 
# more comments on average.
total_ask_comments = 0

for lst in ask_posts:
    total_ask_comments = total_ask_comments + int(lst[4])

avg_ask_comments = total_ask_comments / len(ask_posts)

print('Ask HN total comments and its average')
print(total_ask_comments)
print(avg_ask_comments)
print('\n')

total_show_comments = 0

for lst in show_posts:
    total_show_comments = total_show_comments + int(lst[4])

avg_show_comments = total_show_comments / len(show_posts)

print('Show HN total comments and its average')
print(total_show_comments)
print(avg_show_comments)
```

> Analyzing our code output above, we find that `Ask HN` received more comments on average than `Show HN`.
>
> But obvious `Ask HN` being a two-way interactive post makes the 
> reader of the post feel importance and being involved in it as 
> a part of the on-going conversation. So the reader plays his 
> part by giving comments. Whereas the `Show HN` is just a one 
> way interactive post. Here only the author of the post holds 
> the importance by showcasing what he wants to the reader 
> through his post.
>
> Thus our conclusion is post that makes the reader feel 
> importance holds popularity, as these posts are the post for 
> readers only.

```python
# focus on what matters - Ask HN post
# code to analyse whether the time of the day when the post was 
# created plays a part in its popularity
```

```python
import datetime as dt
```

```python
result_list = []

for lst in ask_posts:
    result_list.append([lst[6] , int(lst[4])])
    
print(result_list)
```

```python
counts_by_hour = {}
comments_by_hour = {}

for lst in result_list:
    print(lst[0])
    datetime_obj = dt.datetime.strptime(lst[0], '%m/%d/%Y %H:%M')
    print(datetime_obj)
    hour = datetime_obj.strftime('%H')
    print(hour)
    if hour not in counts_by_hour:
        counts_by_hour[hour] = 1
        comments_by_hour[hour] = lst[1]
    else:
        counts_by_hour[hour] += 1
        comments_by_hour[hour] += lst[1]
        
print(counts_by_hour)
print(comments_by_hour)
    
```

```python
# use the above information achieved as output of our code to
# calculate the average number of comments for posts created
# during each hour of the day.
```

```python
avg_by_hour = []

for lst in counts_by_hour:
    if lst in comments_by_hour:
        avg_by_hour.append([lst, (comments_by_hour[lst] / counts_by_hour[lst])])

print(avg_by_hour)
```

```python
swap_avg_by_hour = [] 

for lst in avg_by_hour:
    swap_avg_by_hour.append([lst[1], lst[0]])
    
print(swap_avg_by_hour)
```

```python
sorted_swap = sorted(swap_avg_by_hour, reverse = True)
print(sorted_swap[:5])
```

```python
outputsentence = '{hr}: {avg:.2f} average comments per post'

for lst in sorted_swap[:5]:
    hour = dt.datetime.strptime(lst[1], '%H').strftime('%H:%M')
    comment_avg = lst[0]
    print(outputsentence.format(hr = hour, avg = comment_avg))
```

```python
from pytz import all_timezones

for lst in all_timezones:
    if 'Asia' in lst:
        print(lst)
```

```python
from pytz import timezone
```

```python
outputsentence = '{hr} in Eastern Time in the US: {avg:.2f} average comments per post'
outputsentencea = '{hr} in IST: {avg:.2f} average comments per post'

for lst in sorted_swap[:5]:
    hour = dt.datetime.strptime(lst[1], '%H').strftime('%H:%M')
    comment_avg = lst[0]
    print(outputsentence.format(hr = hour, avg = comment_avg))
    est_time = (timezone('US/Eastern').localize(dt.datetime.strptime(hour, '%H:%M')))
    ist_time = est_time.astimezone(timezone('Asia/Calcutta')).strftime('%H:%M')
    print(outputsentencea.format(hr = ist_time, avg = comment_avg))
```

> The output of our code gives us a conclusion that the post on 
> `Hacker News` achieve more comments and become popular if it is 
> of `Ask HN` type and created at a time frame of `3:00pm EST`, 
> Eastern Time in the US or `01:49am IST`.
